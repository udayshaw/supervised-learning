```python
df['class_e']=np.where(df['class']=='e',1,0)
df['bruises_f']=np.where(df['bruises']=='f',1,0)
df['gill-attachment_f']=np.where(df['gill-attachment']=='f',1,0)
df['gill-spacing_c']=np.where(df['gill-spacing']=='c',1,0)
df['gill-size_b']=np.where(df['gill-size']=='b',1,0)
df['stalk-shape_t']=np.where(df['stalk-shape']=='t',1,0)
df['veil-type_p']=np.where(df['veil-type']=='p',1,0)

cap_shape = pd.get_dummies(df['cap-shape'], prefix='cap_shape')
df=pd.concat([df, cap_shape], axis=1)
cap_surface = pd.get_dummies(df['cap-surface'], prefix='cap_surface')
df=pd.concat([df, cap_surface], axis=1)
cap_color = pd.get_dummies(df['cap-color'], prefix='cap_color')
df=pd.concat([df, cap_color], axis=1)
odor = pd.get_dummies(df['odor'], prefix='odor')
df=pd.concat([df, odor], axis=1)
gill_color = pd.get_dummies(df['gill-color'], prefix='gill_color')
df=pd.concat([df, gill_color], axis=1)
stalk_root = pd.get_dummies(df['stalk-root'], prefix='stalk_root')
df=pd.concat([df, stalk_root], axis=1)
ssar = pd.get_dummies(df['stalk-surface-above-ring'], prefix='ssar')
df=pd.concat([df, ssar], axis=1)
ssbr = pd.get_dummies(df['stalk-surface-below-ring'], prefix='ssbr')
df=pd.concat([df, ssbr], axis=1)
scar = pd.get_dummies(df['stalk-color-above-ring'], prefix='scar')
df=pd.concat([df, scar], axis=1)
scbr = pd.get_dummies(df['stalk-color-below-ring'], prefix='scbr')
df=pd.concat([df, scbr], axis=1)
veil_color = pd.get_dummies(df['veil-color'], prefix='veil_color')
df=pd.concat([df, veil_color], axis=1)
ring_number = pd.get_dummies(df['ring-number'], prefix='ring_number')
df=pd.concat([df, ring_number], axis=1)
ring_type = pd.get_dummies(df['ring-type'], prefix='ring_type')
df=pd.concat([df, ring_type], axis=1)
spc = pd.get_dummies(df['spore-print-color'], prefix='spc')
df=pd.concat([df, spc], axis=1)
population = pd.get_dummies(df['population'], prefix='population')
df=pd.concat([df, population], axis=1)
habitat = pd.get_dummies(df['habitat'], prefix='habitat')
df=pd.concat([df, habitat], axis=1)
```

```python
df=df.drop(col_list, axis=1)
df
```

```python
X=df.drop(["class_e"], axis=1)
y=df['class_e']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
Maximum_Depth=[1, 2, 3, 4, 5, 6, 7 ]
Purity_Threshold= [0.1, 0.2, 0.6, 0.7, 0.8, 0.9]
#Size_Threshold: Very low
```

```python
output=[]
for i in Maximum_Depth:
    for j in Purity_Threshold:
        data={}
        data["Maximum Depth"]=i
        data["Purity Threshold"]=j

        classifier = DecisionTreeClassifier(criterion='entropy',
            splitter='random',
            max_depth=i,
            min_impurity_decrease=1-j   
        )
        #since for impurity values greater than 0.2, the accuracy value is same for all depths 
        #so, I have taken (1-impurity) to have some variance in accuracy value
        classifier.fit(X_train, y_train)

        #train accuracy
        y_pred_train = classifier.predict(X_train)
        accuracy_train = accuracy_score(y_train, y_pred_train)
        data["accuracy on train set"]=accuracy_train

        train_mse = mean_squared_error(y_train, y_pred_train)
        data["MSE on train set"]=train_mse

        #test accuracy
        y_pred = classifier.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        data["accuracy on test set"]=accuracy

        test_mse = mean_squared_error(y_test, y_pred)
        data["MSE on test set"]=test_mse
        output.append(data)
        #print(data)
```

```python
pd.DataFrame(output).sort_values(["accuracy on test set"], ascending=False)
```

```python
classifier = DecisionTreeClassifier(criterion='entropy',
            splitter='best',
            max_depth=2,
            min_impurity_decrease=1-0.8)
classifier.fit(X_train, y_train)

text_representation = tree.export_text(classifier, feature_names=list(X_train.columns))
print(text_representation)
```

<p><b>The best combination seems to be depth 2 and impurity 0.8, because any increase in these values does not change the output much</b></p>
<ul>
<li>If odor_none value is less than 0.5 and there are no bruises with value less than 0.5 than the mushroom is edible</li>
<li>If odor_none value is less than 0.5 and there are no bruises with value greater than 0.5 than the mushroom is poisonous</li>
<li>If odor_none value is less than 0.5 than poisonous</li>
</ul>
